## Exercise 01

Write code to read data from: https://data.giss.nasa.gov/gistemp/tabledata_v3/GLB.Ts+dSST.txt . 

Keep all monthly temperatures (from Jan to Dec) between 1980 and 2015 ONLY (ignore any other data), then:

- Find minimum temperature in Jan between 1980 - 2015
- Find average temperature in each year between 1980 - 2015
- Find temperature in Jan 2000
- Find temperature in Jun 2009

```python
import requests
import pandas as pd
import numpy as np
```

```python
#extract and clean data
url = "https://data.giss.nasa.gov/gistemp/tabledata_v3/GLB.Ts+dSST.txt"
df = pd.read_csv(url, skiprows=7, delim_whitespace=True, header=None)
#remove the last few lines of comments
df=df[:-5]
#get only 1980-2015
df=df[104:-4]
#remove the last few cols that includes irrelevant data 
df=df.iloc[:,:-7]
#make months as header
df = pd.DataFrame(df.values[1:], columns=df.iloc[2])
#delete rows with year and months
df = df[df.iloc[:,0]!='Year']
df = df.set_index('Year')
df = df.astype(int)
df
```

```python
# Find minimum temperature in Jan between 1980 - 2015
# Your code here
jan = df['Jan']
jan.loc[jan==min(jan)]
```

```python
# Find average temperature in each year between 1980 - 2015
# Your code here
#df.sum().sum() /(len(df)*12)

#in the form of (year, average temperature)
[(1980+i, sum(df.iloc[i]/12)) for i in range(len(df))]
```

```python
# Find temperature in Jan 2000
# Your code here
print('Average temperature in Jan 2000: '+ str(jan['2000']))
```

```python
# Find temperature in Jun 2009
# Your code here
jun = df['Jun']
print('Average temperature in Jun 2009: '+ str(jun['2009']))
```

`answers.txt` contains students answers of exam questions (each line contains student_id, then answers made by that student)

- The first number is the student’s ID number. The following 25 letters are the student responses to the exam. If there is no letter following a comma, this means the student skipped answering the question.

- Write code to read the data from the file, compare students answers to answers key (1 point for each correct answer), then find: 

    - The grade for each student.
    - Id of student of the highest score.
    - Id of student of the lowest score.

`answerkey = "B,A,D,D,C,B,D,A,C,C,D,B,A,B,A,C,B,D,A,C,A,A,B,D,D"`

```python
answerkey = "B,A,D,D,C,B,D,A,C,C,D,B,A,B,A,C,B,D,A,C,A,A,B,D,D"
key = answerkey.split(',')
ans = pd.read_csv('answers.txt', names = [i for i in range(len(key))])
ans
```

```python
# The grade for each student
# Your code here
def grade(l):
    a = np.zeros(len(l))
    for i in range(len(l)):
        a[i] = (l[i]==key[i])
    return sum(a)

#ans.iloc[0][0]==key[0]
#grade(ans.iloc[2])
grades = pd.DataFrame([grade(ans.iloc[i]) for i in range(len(ans))])
grades = grades.set_index(ans.index)
grades
```

```python
# Id of student of the highest score
# Your code here
grades.idxmax()
```

```python
# Id of student of the lowest score
# Your code here
grades.idxmin()
```
