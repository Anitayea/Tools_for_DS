```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
entire_list = []
search = ['data+science', 'machine+learning', 'python+programming', 'java', 'computer+science']
url_list = ['https://openlibrary.org/search?q='+i+'&mode=everything' for i in search]
for suburl in url_list:
  main_soup = BeautifulSoup(requests.get(suburl).text)
  prev_url = 'https://openlibrary.org'
  link = [prev_url + a['href'] for a in main_soup.find_all('a', {'class': 'results'})]
  link = [i for i in link if 'authors' not in i][0:5]
  cate_index = 0
  for i in link:
    category = search[cate_index]

    soup = BeautifulSoup(requests.get(i).text)

    titles = soup.find_all('h1', {'class': 'work-title'})
    if len(titles) > 0:
      title = titles[0].text
    else:
      title = "unknown"
    
    dates = soup.find_all('span', {'class': "first-published-date"})
    if len(dates) > 0:
      date = dates[0]['title'][-4:]
    else:
      date = "unknown"

    authors = soup.find_all('h2', {'class': "edition-byline"})
    if len(authors) > 0:
      author = authors[0].text.replace('\n', '')[4:]
    else:
      author = 'unknown'

    language = soup.find_all('a', {'href': "/languages/eng"})
    if len(language) > 0:
      lan = language[0].text
    else:
      lan = "unknown"

    list_store = [category, title, date, author, lan]
    #list_store = [i for i in list_store if i != '']
    entire_list.append(list_store)
  cate_index += 1
entire_list[0:5]
```

```python
import pandas as pd

#convert to Dataframe specifying columns names
df = pd.DataFrame(entire_list, columns=['Category','Title','Publish_Time','Author', 'Language'])
df
```

```python
#3
class mystr(str):
    def replace_all(self, ini_list, new_list):
        if len(ini_list) != len(new_list):
            raise ValueError("Lists does not have same length")

        new_str = ''
        for char in self:
            if char in ini_list:
                new_str += new_list[ini_list.index(char)]
            else:
                new_str += char
        return new_str

    def upper_first_last(self):
        if len(self) <= 1:
            return self.upper()
        else:
            return self[0].upper() + self[1:-1] + self[-1].upper()

    def ispalindrome(self):
        return self == self[::-1]

x=mystr('welcome everyone.. today is Tuesday')
print(x.replace_all(['e','o'],['a','u']))
print(mystr.upper_first_last('welcome'))
print(mystr.ispalindrome('welcome'))
print(mystr.ispalindrome('weaew'))
```
